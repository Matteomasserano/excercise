```python
import pandas as pd
import numpy as np
import os 
```

```python
lordo = pd.read_csv('/Users/matteomasserano/Desktop/Data Analysis/Boolean/Progetto finale/Notebook/Salari/Lordo/Total_lordo',index_col=0)
netto = pd.read_csv('/Users/matteomasserano/Desktop/Data Analysis/Boolean/Progetto finale/Notebook/Salari/Netto/Total_netto', index_col=0)
```

```python
netto.head()
```

```python
lordo.head()
```

```python
total = netto.merge(lordo, on=['Year', 'Number'])
total.head()
```

```python
total = total[['Player', 'Age', 'Country', 'Position', 'Club', 'Net Week', 'Net Year', 'Gross Week', 'Gross Year', 'Year']]
```

```python
total.head()
```

```python
total.Position.unique()
```

```python
total['Position'] = total['Position'].replace('K', 'Goalkeeper')
total['Position'] = total['Position'].replace('D', 'Defender')
total['Position'] = total['Position'].replace('M', 'Midfielder')
total['Position'] = total['Position'].replace('F', 'Forward')
```

```python
total
```

```python
seasons = [2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
```

```python
total['Net Week'] = total['Net Week'].str.replace(',', '')
```

```python
total['Gross Week'] = total['Gross Week'].str.replace(',', '')
total['Net Year'] = total['Net Year'].str.replace(',', '')
total['Gross Year'] = total['Gross Year'].str.replace(',', '')
```

```python
total.to_csv('/Users/matteomasserano/Desktop/Data Analysis/Boolean/Progetto finale/Notebook/Salari/Totale/Salari totale.csv')
```

```python
for season in seasons:

    df = total.loc[(total['Year']==season)].copy()

    df.to_csv(f'/Users/matteomasserano/Desktop/Data Analysis/Boolean/Progetto finale/Notebook/Salari/Totale/{season}.csv')
```
